```python
import numpy as np
import pandas as pd
```

```python
# テストデータの生成
ts = pd.date_range( '11/1/2019', periods=120, freq='3min' )
df = pd.DataFrame( {"value":np.arange(0,len(ts))}, index=ts )
df["value"] += 1
df.head()
```

```python
# indexをカラムに戻して、カラム名を変更する
df = df.reset_index()
df = df.rename( columns={'index': 'time'} )
df.head()
```

```python
# 処理しやすいように、時間帯カラム列を追加する
#df['begin'] = df['time'].dt.round("H") # 四捨五入的な丸め？
df['begin'] = df['time'].apply( lambda x : x.replace( minute=0, second=0, microsecond=0 ) )
df.head( 30 )
```

```python
# タイムスタンプまでの、時間帯ごとの発生回数をカウント、カウント列を追加
#df.apply( lambda x : print(x['time']), axis=1 )
df['count'] = df.apply( lambda x : len( df[ (x['begin']<=df['time']) & (df['time']<=x['time']) ] ), axis=1 )
df.head(30)
```
